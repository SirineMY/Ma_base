```python
import pandas as pd
import mysql.connector as mysql
```

```python
DC = pd.read_excel('donnees_communes.xlsx')
DC
```

```python
commune = DC[['CODDEP','COM','PTOT', 'CODCOM']]
```

```python
commune["id_commune"] = commune["CODDEP"].astype(str) + commune["CODCOM"].map("{:03}".format)

communes = commune[['CODDEP','COM','PTOT','id_commune']]
communes
```

```python
communes.dropna()
```

```python
db = mysql.connect(
    host = "localhost",
    user = "root",
    passwd = "",
    database = "db_immo"
)

cursor = db.cursor()

sql = "INSERT INTO commune(Id_dep,Nom_commune,Population,ID_com) VALUES(%s,%s,%s,%s)"
for i,row in communes.iterrows():
    cursor.execute(sql, tuple(row))    
    db.commit()

db.close()
```
