```python
import pandas as pd
import mysql.connector as mysql
```

```python
vf = pd.read_excel('Valeurs_foncières.xlsx')
```

```python
vf = vf[['No voie', 'Type de voie', 'Voie', 'Nombre pieces principales', 'Surface Carrez du 1er lot', 
         'Surface reelle bati', 'Type local', 'Code departement', 'Code commune']]
```

```python
vf["id_code"] = vf['Code departement'].map("{:0>2}".format) + vf['Code commune'].map("{:03}".format)
```

```python
vf['id'] = vf.index
```

```python
bien = vf.astype(object).where(pd.notnull(vf), None)
```

```python
bien = bien[['id', 'No voie', 'Type de voie', 'Voie', 'Nombre pieces principales', 'Surface Carrez du 1er lot',
             'Surface reelle bati', 'Type local', 'id_code']]
bien
```

```python
db = mysql.connect(
    host = "localhost",
    user = "root",
    passwd = "",
    database = "db_immo"
)

cursor = db.cursor()

sql = "INSERT INTO bien(Id_bien, No_voie,Type_voie, Voie, Total_piece, Surface_carrez,Surface_local, Type_local, id_codedep_codecommune) VALUES(%s,%s,%s,%s,%s,%s,%s,%s,%s)"
for i,row in bien.iterrows():
    cursor.execute(sql, tuple(row))    
    db.commit()

db.close()
```
