```python
import pandas as pd
import mysql.connector as mysql
```

```python
ref_geo = pd.read_excel('fr-esr-referentiel-geographique.xlsx')
ref_geo
```

```python
region = ref_geo[['reg_nom','reg_code']]
region
```

```python
region = region.drop_duplicates()
```

```python
db = mysql.connect(
    host = "localhost",
    user = "root",
    passwd = "",
    database = "db_immo"
)

cursor = db.cursor()

sql = "INSERT INTO region(Nom_reg,Id_reg) VALUES(%s,%s)"
for i,row in region.iterrows():
    cursor.execute(sql, tuple(row))    
    db.commit()

db.close()
```
