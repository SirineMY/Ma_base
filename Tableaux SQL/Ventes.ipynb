```python
import pandas as pd
import mysql.connector as mysql
```

```python
valeurs = pd.read_excel('Valeurs_foncières.xlsx')
valeurs
```

```python
ventes = valeurs[['Valeur fonciere']]
ventes
```

```python
ventes['Date'] = valeurs['Date mutation'].astype('str')
```

```python
ventes['id'] = ventes.index
ventes
```

```python
ventes.dropna(inplace=True)
```

```python
ventes.drop_duplicates()
```

```python
db = mysql.connect(
    host = "localhost",
    user = "root",
    passwd = "",
    database = "db_immo"
)

cursor = db.cursor()

sql = "INSERT INTO ventes(Valeurs,Jour,Id_ventes) VALUES(%s,%s,%s)"
for i,row in ventes.iterrows():
    cursor.execute(sql, tuple(row))    
    db.commit()

db.close()
```
